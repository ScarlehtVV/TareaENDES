```python
class Alumno:
    def __init__(self, nombre, apellido1, apellido2, fecha_nacimiento):
        # Atributos iniciales
        self.nombre = nombre
        self.apellido1 = apellido1
        self.apellido2 = apellido2
        self.fecha_nacimiento = fecha_nacimiento
        # Atributos no iniciales
        self.notas = {}

    def atributos(self):
        # Imprime los atributos principales
        print(f"Nombre completo: {self.nombre} {self.apellido1} {self.apellido2}")
        print(f"Fecha de nacimiento: {self.fecha_nacimiento}")

    def mayor_edad(self):
        from datetime import date
        nacimiento = date.fromisoformat(self.fecha_nacimiento)
        edad = (date.today() - nacimiento).days // 365
        return edad >= 18

    def mostrar_notas(self):
        if not self.notas:
            print("No hay notas registradas.")
        else:
            for asignatura, nota in self.notas.items():
                print(f"{asignatura}: {nota}")

    def poner_notas(self, asignatura, nota):
        self.notas[asignatura] = nota
        print(f"Nota de {nota} registrada para {asignatura}.")

# Función para interactuar con el usuario
def crear_alumno_desde_usuario():
    print("Introduce los datos del alumno:")
    nombre = input("Nombre: ")
    apellido1 = input("Primer apellido: ")
    apellido2 = input("Segundo apellido: ")
    fecha_nacimiento = input("Fecha de nacimiento (YYYY-MM-DD): ")
    
    alumno = Alumno(nombre, apellido1, apellido2, fecha_nacimiento)
    
    print("\n¿Quieres añadir notas? (sí/no)")
    respuesta = input().strip().lower()
    
    while respuesta == "sí":
        asignatura = input("Introduce el nombre de la asignatura: ")
        nota = input("Introduce la nota (o deja en blanco si aún no la sabes): ")
        nota = float(nota) if nota else None
        alumno.poner_notas(asignatura, nota)
        
        print("\n¿Quieres añadir otra nota? (sí/no)")
        respuesta = input().strip().lower()
    
    return alumno

def actualizar_notas(alumno):
    print("\n¿Quieres añadir o actualizar notas? (sí/no)")
    respuesta = input().strip().lower()
    
    while respuesta == "sí":
        asignatura = input("Introduce el nombre de la asignatura: ")
        nota = input("Introduce la nueva nota: ")
        nota = float(nota)
        alumno.poner_notas(asignatura, nota)
        
        print("\n¿Quieres añadir o actualizar otra nota? (sí/no)")
        respuesta = input().strip().lower()

# Ejecución del programa
alumno_creado = crear_alumno_desde_usuario()

print("\nDatos del alumno:")
alumno_creado.atributos()

print("\nNotas del alumno:")
alumno_creado.mostrar_notas()

print("\n¿Es mayor de edad?:", alumno_creado.mayor_edad())

# Permitir actualizar notas
actualizar_notas(alumno_creado)

print("\nNotas finales del alumno:")
alumno_creado.mostrar_notas()
```
